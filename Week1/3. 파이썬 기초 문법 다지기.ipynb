## 미션 1: 변수 타입과 형 변환 (Type Casting)

문제 설명: 정수형 변수 a와 문자열 변수 b가 매개변수로 주어질 때, 문자열 b를 정수형으로 변환하여 a와 더한 값을 반환하는 함수를 작성하세요.

```python
a = 10
b = "20"

def add_func(a, b):
    return a + int(b)

result = add_func(a,b)
print(result)
```

## 미션 2: 산술 연산과 비교 (Operators)

문제 설명: 두 정수 num1, num2가 주어질 때, 두 수의 합, 차, 곱, 몫(정수 부분)을 순서대로 담은 리스트를 반환하고, 추가로 num1이 num2보다 큰지 여부(True/False)를 반환하는 함수를 각각 작성하세요.

- 입출력 예시:

- 핵심 코드 (Reference):
    - programmers_python_algorithm/programmers_lv0/arithmetic_operators.py
    - programmers_python_algorithm/programmers_lv0/comparison_operators.py

```python
def verse_func(num1, num2):
    result = [num1+num2, num1-num2, num1*num2, num1//num2]
    print(result)
    print(num1>num2)

num1 = 10
num2 = 3
verse_func(num1, num2)
```

## 미션 3: 조건에 따른 로직 분기 (If-Elif-Else)

문제 설명: 학생의 점수 score가 정수로 주어질 때, 다음 기준에 따라 학점을 문자열로 반환하는 함수를 작성하세요.
- 90점 이상: "A"
- 70점 이상 90점 미만: "B"
- 70점 미만: "C"

- 입출력 예시:
    - 입력: score = 85
    - 출력: "B"
- 핵심 코드 (Reference):
    - programmers_python_algorithm/programmers_lv0/if_elif_else_statement.py
    - programmers_python_algorithm/programmers_lv0/basic_if_else.py

```python
def test_score(score):
    if score >= 90:
        print('A')
    elif score >= 70:
        print('B')
    else:
        print('C')

score = 85
test_score(score)
```

## 미션 4: 범위 반복문 활용 (For Loop & Range)
문제 설명: 정수 n이 주어질 때, 1부터 n까지의 모든 정수의 합을 구하여 반환하세요. for 반복문과 range() 함수를 반드시 사용해야 합니다.

- 입출력 예시:
    - 입력: n = 5
    - 출력: 15 (1 + 2 + 3 + 4 + 5)
- 핵심 코드 (Reference):
    - programmers_python_algorithm/programmers_lv0/for_loop_with_range.py

```python
def int_sum(n):
    result = 0
    for i in range(1,n+1):
        result += i
    print(result)

n = 5
int_sum(n)
```

## 미션 5: 문자열 포맷팅과 인덱싱 (String Formatting)
문제 설명: 이름 name(문자열)과 나이 age(정수)가 주어질 때, f-string을 사용하여 "My name is [name] and I am [age] years old." 형식의 문자열을 반환하고, 별도로 name의 첫 번째 글자만 추출하여 반환하세요.

- 입출력 예시:
    - 입력: name="Alice", age=25
    - 출력: ("My name is Alice and I am 25 years old.", "A")
- 핵심 코드 (Reference):
    - programmers_python_algorithm/programmers_lv0/f_string_formatting.py
    - programmers_python_algorithm/programmers_lv0/basic_string_indexing.py

```python
def func_name_age(name, age):
    print(f'My name is {name} and I am {age} years old.', name[0])

name = "Alice"
age = 25
func_name_age(name, age)
```

## 미션 6: 리스트 다루기 (List Operations)
문제 설명: 빈 리스트를 생성하고, 숫자 1, 2, 3을 순서대로 추가(append)한 뒤, 최종 리스트의 길이(원소 개수)를 반환하는 함수를 작성하세요.

- 입출력 예시:
    - 입력: 없음
    - 출력: 3 (리스트 [1, 2, 3]의 길이)
- 핵심 코드 (Reference):
    - programmers_python_algorithm/programmers_lv0/list_creation_and_indexing.py
    - programmers_python_algorithm/programmers_lv0/list_append_method.py
    - programmers_python_algorithm/programmers_lv0/len_function_usage.py

```python

def count():
    nums = []
    for i in range(3):
        nums.append(i+1)
    return len(nums)

result = count()
print(result)
```
