## 미션 1: 조건에 맞는 데이터 추출
문제 설명: 정수로 이루어진 리스트 numbers가 주어집니다. 이 리스트에서 짝수만을 골라내고, 골라낸 값들을 제곱하여 새로운 리스트를 반환하는 함수를 작성하세요. (가능하면 for 반복문 대신 리스트 컴프리헨션을 사용해 한 줄로 구현해 보세요.)

- 입출력 예시:
    - 입력: numbers = [1, 2, 3, 4, 5, 6]
    - 출력: [4, 16, 36] (2의 제곱, 4의 제곱, 6의 제곱)
    - 입력: numbers = [1, 3, 5]
    - 출력: []
- 핵심 코드 (Reference):
    - programmers_python_algorithm/Week1/list_comprehensions.py
    - programmers_python_algorithm/Week1/list_slicing.py

```python
def even_num(numbers):
    return [i**2 for i in numbers if i%2 == 0]

numbers = [1,2,3,4,5,6]
result =  even_num(numbers)
print(result)

numbers = [1,3,5]
result =  even_num(numbers)
print(result)
```

## 미션 2: 데이터 묶기 (Zip & Unpacking)

문제 설명: 학생들의 이름이 담긴 리스트 names와 점수가 담긴 리스트 scores가 주어집니다. 인덱스가 같은 이름과 점수를 짝지어 {'이름': 점수} 형태의 딕셔너리로 변환하여 반환하세요.

- 입출력 예시:
    - 입력: names = ['철수', '영희', '민수'], scores = [90, 85, 70]
    - 출력: {'철수': 90, '영희': 85, '민수': 70}
- 핵심 코드 (Reference):
    - programmers_python_algorithm/Week1/tuple_unpacking.py
    - codingtest_python/reference/zip.py

```python
def name_score(names, scores):
    return dict(zip(names, scores))

names = ['철수', '영희', '민수']
scores = [90, 85, 70]

result = name_score(names, scores)
result
```

## 미션 3: 중복 제거 및 교집합 찾기

문제 설명: 두 개의 정수 리스트 listA와 listB가 주어집니다. 두 리스트에 공통으로 포함된 숫자들만 찾아서 오름차순으로 정렬된 리스트를 반환하세요. (반복문을 사용한 O(N^2) 비교 대신 set 자료구조의 교집합 연산을 활용하세요.)

- 입출력 예시:
    - 입력: listA = [1, 5, 2, 3], listB = [2, 3, 4, 5, 9]
    - 출력: [2, 3, 5] (2, 3, 5가 양쪽에 모두 존재함)
- 핵심 코드 (Reference):
    - programmers_python_algorithm/Week1/set_operations.py
    - codingtest_python/performance/list_vs_set_in.py

```python
def check_double(listA, listB):
    result = set(listA) & set(listB)
    return sorted(result)


listA = [1, 5, 2, 3]
listB = [2, 3, 4, 5, 9]

result = check_double(listA, listB)
result
```

## 미션 4: 문자열 빈도수 계산기
문제 설명: 문자열 s가 주어졌을 때, 각 문자가 몇 번 등장하는지 세어 {'문자': 등장횟수} 형태의 딕셔너리로 반환하세요. collections 모듈의 기능을 활용하면 더 간단하게 구현할 수 있습니다.
