## 미션 1: Zip을 활용한 데이터 매핑
문제 설명: 과목명이 담긴 리스트 subjects와 해당 과목의 점수가 담긴 리스트 scores가 주어집니다. 

for 문으로 하나씩 채워 넣는 방식 대신, zip 함수를 활용하여 {'과목명': 점수} 형태의 딕셔너리를 한 줄로 생성하세요.

- 입출력 예시:
    - 입력: subjects = ['Math', 'Sci'], scores = [90, 80]
    - 출력: {'Math': 90, 'Sci': 80}
- 핵심 코드(Reference):
    - codingtest_python/reference/zip.py
    - programmers_python_algorithm/programmers_lv1/zip_for_parallel_iteration.py

````python
def class_score(subjects, scores):
    result = {}
    for sub, score in zip(subjects, scores):
        result[sub] = score
    print(result)

subjects = ['Math', 'Sci']
scores = [90, 80]

class_score(subjects, scores)
````

## 미션 2: Enumerate를 활용한 조건부 인덱스 찾기
문제 설명: 정수 리스트 nums와 목표값 target이 주어집니다. 리스트를 순회하며 target과 일치하는 모든 요소의 인덱스를 리스트에 담아 반환하세요. range(len(nums)) 대신 enumerate를 사용하여 구현해야 합니다.

- 입출력 예시:
    - 입력: nums = [1, 2, 3, 2, 1], target = 2
    - 출력: [1, 3]
- 핵심 코드(Reference):
    - programmers_python_algorithm/programmers_lv1/for_loop_with_enumerate.py

````python
def collect_target(nums, target):
    result = [idx for idx, n in enumerate(nums) if n == target]
    print(result)

nums = [1,2,3,2,1]
target = 2

collect_target(nums, target)
````

## 미션 4: Ternary Operator를 이용한 인라인 조건문
문제 설명: 정수 n이 매개변수로 주어집니다. n이 0보다 크면 "Positive", 0이면 "Zero", 0보다 작으면 "Negative"를 반환하는 함수를 작성하세요. 일반적인 if-else 블록 대신, 한 줄로 표현 가능한 삼항 연산자(Ternary Operator) 형태를 중첩하여 작성해 보세요.

- 입출력 예시:
    - 입력: n = -5 → 출력: "Negative"
- 핵심 코드(Reference):
    - programmers_python_algorithm/programmers_lv1/conditional_expression_ternary.py

````python
def ternary_operator(n):
    result = "Positive" if n > 0 else ("Zero" if n == 0 else "Negative")
    print(result)

n = -5
ternary_operator(n)
````

## 미션 5: 별표(*)를 활용한 가변 언패킹
문제 설명: 리스트 data에는 시험 점수들이 들어있습니다. 가장 낮은 점수와 가장 높은 점수를 제외한 나머지 점수들의 평균을 구하기 위해, 언패킹(*)을 사용하여 min_score, mid_scores(리스트), max_score로 데이터를 분리하세요. (단, 데이터는 정렬되어 있다고 가정합니다.)

- 입출력 예시:
    - 입력: data = [50, 70, 80, 90, 100]
    - 결과: min=50, mid=[70, 80, 90], max=100
- 핵심 코드(Reference):
    - programmers_python_algorithm/Week1/tuple_unpacking.py

````python
def split_list(data):
    min_data, *mid_datas, max_data = data
    return min_data, mid_datas, max_data

data = [50, 70, 80, 90, 100]
min_data, mid_datas, max_data = split_list(data)
min_data, mid_datas, max_data
````

````python
# 다른 예시
def split_list(data):
    a, b, *c, d,  = data
    return a, b, c, d
    

data = [50, 70, 80, 90, 100, 110, 120]
a, b, c, d = split_list(data)
a, b, c, d
````

## 미션 6: In 연산자를 활용한 존재 여부 확인
문제 설명: 문자열 s가 주어질 때, 해당 문자열에 모음(a, e, i, o, u)이 포함되어 있는지 여부를 판별하는 효율적인 코드를 작성하세요. 여러 번의 or 조건문 대신 in 연산자와 집합(Set)을 활용하여 파이썬스럽게 구현하세요.

- 입출력 예시:
    - 입력: s = "sky" → 출력: False
    - 입력: s = "apple" → 출력: True
- 핵심 코드(Reference):
    - programmers_python_algorithm/programmers_lv0/in_operator_usage.py
    - codingtest_python/reference/in.py

````python
def check(s):
    str_list = ['a', 'e', 'i', 'o', 'u']
    result = any(i in str_list for i in set(s))
    print(result)

s = "sky"
check(s)
````

````python
s = "apple"
check(s)
````

any()의 역할 (기능)

- any(iterable)은 이터러블(반복 가능한 것)을 받아서, 원소들 중 하나라도 True가 있으면 True 전부 False면 False 를 반환


“any 없으면 안 되나?”
- 없으면 여러 개의 True/False 결과를 하나의 결론(True/False)로 합칠 방법이 없습니다.
- 즉, generator는 True/False를 “여러 개 만들어낼 뿐” any()가 그걸 “하나라도 True냐?”로 요약(집계) 해줍니다.
- any() 없이 하려면 결국 이런 식으로 OR 누적을 직접 해야 해요:

```
result = False
for i in set(s):
    result = result or (i in str_list)
```

## 미션 7: Map과 Lambda를 활용한 일괄 변환

````python
def change(str_nums):
    result = list(map(int, str_nums))
    print(result)

str_nums = ["1", "2", "3"]
change(str_nums)
````

## 미션 9: Key를 이용한 커스텀 정렬 (문자열 길이)
문제 설명: 단어 리스트 words가 주어집니다. 이 단어들을 단어의 길이를 기준으로 오름차순 정렬하세요. 길이가 같다면 사전순으로 정렬해야 합니다. sort 메서드의 key 인자에 lambda를 활용하세요.

- 입출력 예시:
    - 입력: words = ["apple", "no", "banana", "cat"]
    - 출력: ["no", "cat", "apple", "banana"]
- 핵심 코드(Reference):
    - programmers_python_algorithm/DataStructure/list_sorting.py

````python
def sort_words(words):
    words.sort(key=lambda x: (len(x), x))
    print(words)

words = ["apple", "no", "banana", "cat"]
sort_words(words)
````

## 미션 10: Set 차집합을 이용한 유일 원소 찾기
문제 설명: 두 리스트 A와 B가 주어집니다. A 리스트에는 들어있지만 B 리스트에는 들어있지 않은 원소들만 골라 리스트로 반환하세요. 중복은 제거되어야 하며, 집합(Set)의 차집합 연산을 활용하세요.

- 입출력 예시:
    - 입력: A = [1, 2, 2, 3], B = [3, 4]
    - 출력: [1, 2]
- 핵심 코드(Reference):
    - programmers_python_algorithm/Week1/set_operations.py
    - programmers_python_algorithm/DataStructure/set_basics.py

````python
def unique(A, B):
    result = list(set(A) - set(B))
    print(result)

A = [1, 2, 2, 3]
B = [3, 4]
unique(A, B)
````
